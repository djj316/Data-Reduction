```python
from ucimlrepo import fetch_ucirepo  
from sklearn.decomposition import PCA
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.sans-serif'] = ['SimHei']  # 设置中文字体为黑体
matplotlib.rcParams['axes.unicode_minus'] = False     # 解决负号 "-" 显示成方块的问题


# fetch dataset 
iris = fetch_ucirepo(id=53) 
  
# data (as pandas dataframes) 
x = iris.data.features #包括每个变量
y = iris.data.targets #具体分类
z = iris.data

# print(x)
# print(y)
# print(z)

pca = PCA(n_components=2)
x_reduced = pca.fit_transform(x)

# 转换成DataFrame格式
x_pca = pd.DataFrame(x_reduced, columns=['PC1', 'PC2'])

# 合并标签以便后续可视化
df_pca = pd.concat([x_pca, y.reset_index(drop=True)], axis=1)

# ---------------------
# 可视化降维结果
# ---------------------
plt.figure(figsize=(8,6))
for label in df_pca['class'].unique():
    plt.scatter(df_pca[df_pca['class'] == label]['PC1'],
                df_pca[df_pca['class'] == label]['PC2'],
                label=label)
plt.xlabel('主成分1')
plt.ylabel('主成分2')
plt.title('Iris数据集的PCA降维结果')
plt.legend()
plt.grid(True)
plt.show()

# ---------------------
# 查看各主成分的解释方差比（信息保留比例）
# ---------------------
print("每个主成分解释的方差占比（information retained）：")
print(pca.explained_variance_ratio_)
```

```python
from ucimlrepo import fetch_ucirepo  
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from sklearn.preprocessing import LabelEncoder

# 设置中文字体
matplotlib.rcParams['font.sans-serif'] = ['SimHei']  # 黑体
matplotlib.rcParams['axes.unicode_minus'] = False     # 负号正常显示

# 载入数据
iris = fetch_ucirepo(id=53)  
x = iris.data.features  
y = iris.data.targets  

# 标签编码为整数（LDA需要数值型）
le = LabelEncoder()
y_encoded = le.fit_transform(y['class'])

# 执行LDA降维
lda = LDA(n_components=2)
x_lda = lda.fit_transform(x, y_encoded)

# 转换为DataFrame方便可视化
x_lda_df = pd.DataFrame(x_lda, columns=['判别成分1', '判别成分2'])
x_lda_df['类别'] = y['class'].values

# 绘图
plt.figure(figsize=(8,6))
for label in x_lda_df['类别'].unique():
    plt.scatter(x_lda_df[x_lda_df['类别'] == label]['判别成分1'],
                x_lda_df[x_lda_df['类别'] == label]['判别成分2'],
                label=label)
plt.xlabel('判别成分1')
plt.ylabel('判别成分2')
plt.title('LDA降维：Iris数据集')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris
from ucimlrepo import fetch_ucirepo 

# 加载 Iris 数据集作为示例
iris = load_iris()
X = iris.data  # 特征矩阵
y = iris.target  # 标签

print(X)
print(y)
# 1. 划分数据集为训练集和验证集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 2. 使用训练集进行 LDA 数据规约
lda = LinearDiscriminantAnalysis(n_components=2)  # 降到2维
X_train_lda = lda.fit_transform(X_train, y_train)

# 3. 在降维后的训练集上训练分类模型
knn_lda = KNeighborsClassifier(n_neighbors=3)
knn_lda.fit(X_train_lda, y_train)

# 4. 在原始训练集上训练分类模型
knn_no_lda = KNeighborsClassifier(n_neighbors=3)
knn_no_lda.fit(X_train, y_train)

# 5. 在验证集上评估两个模型的准确度
X_test_lda = lda.transform(X_test)  # 使用训练好的 LDA 模型将测试集数据映射到 2 维
y_pred_lda = knn_lda.predict(X_test_lda)
accuracy_lda = accuracy_score(y_test, y_pred_lda)

y_pred_no_lda = knn_no_lda.predict(X_test)
accuracy_no_lda = accuracy_score(y_test, y_pred_no_lda)

# 输出结果
print(f"Accuracy with LDA: {accuracy_lda:.2f}")
print(f"Accuracy without LDA: {accuracy_no_lda:.2f}")

# 6. 可视化降维后的数据
plt.figure(figsize=(8, 6))
plt.scatter(X_test_lda[:, 0], X_test_lda[:, 1], c=y_test, cmap='viridis', edgecolor='k', s=100)
plt.title("LDA: 2D Projection of the Iris Dataset (Test Set)")
plt.xlabel("LDA Component 1")
plt.ylabel("LDA Component 2")
plt.colorbar(label='Class Label')
plt.show()
```

```python
from ucimlrepo import fetch_ucirepo
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from matplotlib.colors import Normalize
norm = Normalize(vmin=0, vmax=2)  # 确保颜色范围是 0, 1, 2

# 1. 获取 Wine Quality 数据集
wine_quality = fetch_ucirepo(id=186)

# 获取特征和标签
X = wine_quality.data.features#11维
y = wine_quality.data.targets
# print(y)

# 2. 确保 X 是二维特征矩阵，y 是一维标签数组
X = np.array(X)
y = np.array(y).flatten()

# print(X)
# print(y)
# 3. 将质量评分转换为三类标签：低质量（y < 3）、中质量（3 <= y < 6）、高质量（6 <= y <= 10）

y = np.digitize(y, bins=[3, 6], right=True)  # digitize 输出 1、2、3
y = y - 1  # 转换为 0、1、2

# print(X)
# print(y)
# 4. 划分训练集与测试集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# ==========================
# 5. 标准化数据用于 PCA
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 6. 找到最佳 PCA 维度数（保留 95% 方差）
pca_full = PCA()
pca_full.fit(X_train_scaled)
cumulative_variance = np.cumsum(pca_full.explained_variance_ratio_)
n_components_95 = np.argmax(cumulative_variance >= 0.95) + 1
print(f"保留 95% 方差所需 PCA 维度数: {n_components_95}")

# 7. 使用 PCA 降维
pca = PCA(n_components=n_components_95)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

# 8. 使用 LDA 降维
lda = LinearDiscriminantAnalysis(n_components=2)
X_train_lda = lda.fit_transform(X_train, y_train)
X_test_lda = lda.transform(X_test)

# 9. 训练 KNN 模型
knn_lda = KNeighborsClassifier(n_neighbors=3).fit(X_train_lda, y_train)
knn_pca = KNeighborsClassifier(n_neighbors=3).fit(X_train_pca, y_train)
knn_raw = KNeighborsClassifier(n_neighbors=3).fit(X_train, y_train)

# 10. 预测与评估
accuracy_lda = accuracy_score(y_test, knn_lda.predict(X_test_lda))
accuracy_pca = accuracy_score(y_test, knn_pca.predict(X_test_pca))
accuracy_raw = accuracy_score(y_test, knn_raw.predict(X_test))

# 11. 打印准确率
print(f"Accuracy with LDA: {accuracy_lda:.2f}")
print(f"Accuracy with PCA (自动维度={n_components_95}): {accuracy_pca:.2f}")
print(f"Accuracy without LDA or PCA: {accuracy_raw:.2f}")

# 12. 可视化 LDA 和 PCA 降维结果
plt.figure(figsize=(14, 6))

# LDA 可视化
plt.subplot(1, 2, 1)
scatter_lda = plt.scatter(X_test_lda[:, 0], X_test_lda[:, 1],
                          c=y_test, cmap='viridis', norm=norm,
                          edgecolor='k', s=80)
plt.title("LDA: 2D Projection")
plt.xlabel("LDA Component 1")
plt.ylabel("LDA Component 2")
cbar_lda = plt.colorbar(scatter_lda, ticks=[0, 1, 2])
cbar_lda.ax.set_yticklabels(['Low', 'Medium', 'High'])

# PCA 可视化（前2维）
plt.subplot(1, 2, 2)
scatter_pca = plt.scatter(X_test_pca[:, 0], X_test_pca[:, 1],
                          c=y_test, cmap='viridis', norm=norm,
                          edgecolor='k', s=80)
plt.title("PCA: First 2 Components")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
cbar_pca = plt.colorbar(scatter_pca, ticks=[0, 1, 2])
cbar_pca.ax.set_yticklabels(['Low', 'Medium', 'High'])

plt.tight_layout()
plt.show()

# 13. 画出 PCA 累计解释方差图
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o')
plt.axhline(y=0.95, color='r', linestyle='--', label='95% Variance')
plt.xlabel("PCA 维度数")
plt.ylabel("累计解释方差比")
plt.title("PCA 累计解释方差图")
plt.grid(True)
plt.legend()
plt.show()
```
